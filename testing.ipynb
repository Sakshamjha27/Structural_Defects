```python
import google.generativeai as genai 
```

```python
from PIL import Image
```

```python
import os
```

```python
defect_image = Image.open('istockphoto-1083303192-1024x1024.jpg')
```

```python
defect_image
```

```python
gemini_api_key = os.getenv('TestProject1')
genai.configure(api_key=gemini_api_key)
model = genai.GenerativeModel('gemini-2.5-flash-lite')
```

```python
prompt = 'Idenfity the structural defect in th image and suggest how it will be fixed'
```

```python
response = model.generate_content([prompt,defect_image], 
                                  generation_config={'temperature':0.7}) 
```

```python
print(response.text)
```
